## Microsoft: Dat101x Introduction to Data Science
### Module 2: Data Analysis Fundamentals -> Lab Check

````python
print("NOTE: Lab is actually done in MS Excel (not directly in Python as shown here)")

print("Question 1:")
print("\tWhat is the total sum of revenue for August?")
df_august = df[df.Date.dt.month == 8]
s_august_revenue = df_august["Price"] * df_august["Sales"]
august_revenue_sum = s_august_revenue.sum()
print("\tTotal sum of revenue for August:", august_revenue_sum)

print("Question 2:")
print("\tWhat was the temperature on the hottest Saturday in July?")
# NOTE: The boolean operators (and, or) require boolean operands
#       For pandas these are ambiguous so use bitwise operators | for or and & for and
df_july_saturdays = df[(df.Date.dt.month == 7) & (df.Day == "Saturday")]
s_july_saturdays_temperature = df_july_saturdays["Temperature"]
july_saturdays_temperature_max = s_july_saturdays_temperature.max()
print("\tTemperature on the hottest Saturday in July:", july_saturdays_temperature_max)

print("Question 3:")
print("\tWhat was the lowest number of flyers distributed in a day during November?")
df_november = df[df.Date.dt.month == 11]
s_november_flyers = df_november["Flyers"]
print("\tLowest number of flyers distributed in a day during November:", s_november_flyers.min())

print("Question 4:")
print("\tOn which day of the week were most flyers distributed?")
df_groupby_day = df[['Day', 'Flyers']].groupby('Day').sum()
df_groupby_day_max_flyers = df_groupby_day[df_groupby_day.Flyers == df_groupby_day.Flyers.max()]
day_of_week_most_flyers = df_groupby_day_max_flyers.index[0]
print("\tDay of the week were most flyers distributed:", day_of_week_most_flyers)

print("Question 5:")
print("\tOn which day of the week were fewest flyers distributed?")
df_groupby_day_min_flyers = df_groupby_day[df_groupby_day.Flyers == df_groupby_day.Flyers.min()]
day_of_week_fewest_flyers = df_groupby_day_min_flyers.index[0]
print("\tDay of the week were fewest flyers distributed:", day_of_week_fewest_flyers)
````

````python
print("NOTE: Lab is actually done in MS Azure Machine Learning Studio (not directly in Python as shown here)")
      
print("NOTE: predictions here are slightly different from answer provided in lab because we use slightly different math!")
print("Question 1:")
print("\tWhat predictions did you obtain for lemonade sales today and tomorrow with the following data:")
print("\t\tDate     Temperature Rainfall Flyers Price")
print("\t\tToday    37.6        0.8      100    0.3")
print("\t\tTomorrow 41.2        1.02     100    0.3")

from sklearn.linear_model import LinearRegression
X = df[['Temperature', 'Rainfall', 'Flyers', 'Price']]
#print(X)
y = df[['Sales']]
#print(y)
model = LinearRegression()  # create the linear regression model
model.fit(X, y)             # train the linear regression model
# Date     Temperature Rainfall Flyers Price
# Today    37.6        0.8      100    0.3
# Tomorrow 41.2        1.02     100    0.3
today_data = np.array([[37.6, 0.8, 100, 0.3]])
today_sales_predicted = model.predict(today_data)
print("Today sales predicted:", today_sales_predicted.flatten()[0], "(lab provided answer is 19.00269119)")    
tomorrow_data = np.array([[41.2, 1.02, 100, 0.3]])
tomorrow_sales_predicted = model.predict(tomorrow_data)
print("Tomorrow sales predicted:", tomorrow_sales_predicted.flatten()[0], "(lab provided answer is 18.75381906)")

print("Question 2")
print("\tWhat two effects does changing the threshold have on the model evaluation metrics?")
print("\tIncreasing the threshold results in fewer false positives but more false negatives.")
print("\tDecreasing the threshold results in fewer false negatives but more false positives.")

print("Question 3")
print("\tTo which cluster was customer 1 (Marva Cardenas) assigned?")
print("\tUsing K‐Means clustering:")
print("\tCustomer 1 (Marva Cardenas) was assigned to", 2)
````

````python
# NOTE: This example needs more work!!!!!!!!!!!!!!

from sklearn.cluster import KMeans
df_customers = pd.read_csv('./data/Customers.csv', encoding = "ISO-8859-1")
df_customers = df_customers[['Age', 'AvgWeeklySales', 'AvgDrinks']]
kmeans = KMeans(n_clusters=4) # create kmeans object for 4 clusters
kmeans.fit(df_customers)
centroids = kmeans.cluster_centers_
print(centroids)
plt.scatter(df_customers['Age'], df_customers['AvgWeeklySales'], c= kmeans.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
plt.show()
````
