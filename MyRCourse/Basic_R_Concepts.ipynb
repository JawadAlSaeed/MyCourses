# Objects in the Environment

```
- ls()      # list objects in environment
- objects() # list objects in environment
- rm()      # remove objects in environment
```

````r
x <- 42
ls()
'---'
ls(all = TRUE)
'---'
objects()
'---'
objects(all = TRUE)
'---'
````

````r
rm(list=ls()) # remove all non-hidden objects from environment
ls()
'---'
ls(all = TRUE)
'---'
````

````r
rm(list=ls())
ls(all = TRUE)
'---'
x <- 42
.x <- 42 # hidden
ls()
'---'
ls(all = TRUE)
'---'
rm(x)
rm(.x)
ls(all = TRUE)
'---'
````

## Accessing Object Information

```
class()      # what kind of object is it (high-level)?
typeof()     # what is the object’s data type (low-level)?
length()     # how long is it?
dim()        # retrieve or set the dimension of an object
names()      # 
dimnames()   # 
attributes() # does it have metadata?
str()        # Compactly display internal structure of an R object
```

````r
x <- 'abc'; cat(x, class(x), '\n')
x <- "xyz"; cat(x, class(x), '\n')
x <- 42; cat(x, class(x), '\n')
x <- 3.141592; cat(x, class(x), '\n')
x <- 42L; cat(x, class(x), '\n')
x <- TRUE; cat(x, class(x), '\n')
x <- FALSE; cat(x, class(x), '\n')
x <- T; cat(x, class(x), '\n')
x <- F; cat(x, class(x), '\n')
x <- 1 > 2; cat(x, class(x), '\n')
x <- 3+4i; cat(x, class(x), '\n')
````
