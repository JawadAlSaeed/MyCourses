```python
import pandas as pd

players_csv   = "https://raw.githubusercontent.com/chadwickbureau/baseballdatabank/master/core/People.csv"
batting_csv   = "https://raw.githubusercontent.com/chadwickbureau/baseballdatabank/master/core/Batting.csv"

players  = pd.read_csv(players_csv)
batting  = pd.read_csv(batting_csv)

display(players.head())
display(batting.head())
```

```python
print(players['playerID'].nunique())  # number of unique playerID in players
print(len(players['playerID']))       # number of rows in players
print(batting['playerID'].nunique())  # number of unique playerID in bating
print(len(batting['playerID']))       # number of rows in batting
```

```python
player_and_batting = pd.merge(batting, players, on='playerID')
display(player_and_batting.head())
```
