```python
import numpy as np
import matplotlib.pyplot as plt
x=np.setdiff1d(np.linspace(0,10,100),[0]) # [0] to remove the zero
plt.plot(x, 1/x, color='r')
plt.title('Functional vs Object Oriented Programming')
plt.plot([0,0], [0, 10], color='k')
plt.plot([0,10], [0, 0], color='k')
plt.text(0.5, 8, "functional programming")
plt.text(4, 1, "object oriented programming")
plt.tick_params(top=False, bottom=False, left=False, right=False, labelleft=False, labelbottom=False)
plt.xlabel("probability of programming errors")
plt.ylabel("learning difficulty")
plt.show()
```
