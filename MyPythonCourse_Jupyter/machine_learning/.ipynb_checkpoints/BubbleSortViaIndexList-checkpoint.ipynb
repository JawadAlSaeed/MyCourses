```python
# Create the algorithm outlined below and implement it with the following specifications:

# mSmallest( L[1..n])
# Pre: L is a list of distinct integer values. n is the number of elements in the list.
# Post: A list [p1, p2, …, pm] has been returned where pj is the position of the jth smallest element of L.
# The original list L is not changed.
#
#    • prompt the user to enter the list of numbers
#    • prompt the user to enter the value for m
#    • print a list of the positions where the m smallest values are located.
# Note: elements entered by the user are unique.

# Note: bubble sort average and worst-case performance is O(n^2)

def mSmallest(aList = None): # bubble sort but using index-index (see: http://python3.codes/2015/08/24/popular-sorting-algorithms)
    if aList is None:
        print("Enter list of numbers seperated with spaces:") # prompt user
        aList = [int(x) for x in input().split()]
    length = len(aList)
    print("length:", length)
    resultList = list(range(length))
    complexityCounter = 0
    changed = True
    while changed:
        changed = False
        for i in range(len(resultList) - 1):
            complexityCounter += 1
            if aList[resultList[i]] > aList[resultList[i+1]]:
                resultList[i], resultList[i+1] = resultList[i+1], resultList[i]
                changed = True
    print("complexityCounter:", complexityCounter)            
    return resultList
    
aList = [61, 48, 92, 8, 17, 2, 63, 5, 7, 14]

resultList = mSmallest(aList)

print("\nIndexes in sorted value order:")
for x in resultList:
    print(x)
    
print("\nSorted values:")
for x in resultList:
    print(aList[x])

print("\nNote that original list has not been changed:")
for x in aList:
    print(x)

print("\nTry it with user input:")
resultList = mSmallest()
print("\nIndexes in sorted value order:")
for x in resultList:
    print(x)
```
