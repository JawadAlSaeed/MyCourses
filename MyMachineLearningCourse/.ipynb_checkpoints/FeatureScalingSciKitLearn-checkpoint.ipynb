## StandardScaler

The ```StandardScaler``` scales each feature so that **mean = 0** with **stdev = 1** (assumes normally distributed data)

$$ \frac{x_i-mean(x)}{stdev(x)}$$

````python
%matplotlib inline

import pandas as pd
import numpy as np
from sklearn import preprocessing
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

matplotlib.style.use('ggplot')
np.random.seed(1)
df = pd.DataFrame({
    'x1': np.random.normal(0, 2, 10000), # normal distribution, mean=0, stdev=2, size=1000
    'x2': np.random.normal(5, 3, 10000), # normal distribution, mean=5, stdev=3, size=1000
    'x3': np.random.normal(-5, 5, 10000) # normal distribution, mean=-5, stdev=5, size=1000
})

scaler = preprocessing.StandardScaler()
scaled_df = scaler.fit_transform(df)
scaled_df = pd.DataFrame(scaled_df, columns=['x1', 'x2', 'x3'])

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(6, 5))

ax1.set_title('Before Scaling')
sns.kdeplot(df['x1'], ax=ax1)
sns.kdeplot(df['x2'], ax=ax1)
sns.kdeplot(df['x3'], ax=ax1)
ax2.set_title('After Standard Scaler')
sns.kdeplot(scaled_df['x1'], ax=ax2)
sns.kdeplot(scaled_df['x2'], ax=ax2)
sns.kdeplot(scaled_df['x3'], ax=ax2)
plt.show()
````

## MinMaxScaler

The ```MinMaxScaler``` scales each feature so that **range** is in [0,1] or [-1 to 1] if negative values (does not assume normally distributed data)

$$ \frac{x_i-mean(x)}{max(x)-min(x)}$$

````python
df = pd.DataFrame({
    # positive skew
    'x1': np.random.chisquare(8, 1000), # degrees-of-freedom = 8, size=1000 (sum of squares of 8 normal distributions)
    # negative skew 
    'x2': np.random.beta(8, 2, 1000) * 40,
    # no skew
    'x3': np.random.normal(50, 3, 1000)
})

scaler = preprocessing.MinMaxScaler()
scaled_df = scaler.fit_transform(df)
scaled_df = pd.DataFrame(scaled_df, columns=['x1', 'x2', 'x3'])

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(6, 5))
ax1.set_title('Before Scaling')
sns.kdeplot(df['x1'], ax=ax1)
sns.kdeplot(df['x2'], ax=ax1)
sns.kdeplot(df['x3'], ax=ax1)
ax2.set_title('After Min-Max Scaling')
sns.kdeplot(scaled_df['x1'], ax=ax2)
sns.kdeplot(scaled_df['x2'], ax=ax2)
sns.kdeplot(scaled_df['x3'], ax=ax2)
plt.show()
````
